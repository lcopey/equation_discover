```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import RNN, Embedding, TextVectorization, StringLookup, GRU, LSTM, Dense
```

```python
import tensorflow_datasets as tfds
```

```python
import numpy as np
```

```python
text = np.array(['+', 'sin', 'x1']).reshape(-1, 1)
tokens = np.unique(text)
```

```python
vectorizer = StringLookup(vocabulary=tokens)
```

```python
vectorizer(text)
```

```python
embedder = Embedding(input_dim=len(vectorizer.get_vocabulary()), output_dim=12)
```

```python
embedder(vectorizer(text))
```

```python
recurrent_layer = LSTM(128)
```

```python
recurrent_layer(embedder(vectorizer(text)))
```

```python
dense = Dense(len(tokens), activation='softmax')
```

```python
dense(recurrent_layer(embedder(vectorizer(text))))
```
